```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lux
import csv
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
# importing data
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(url)
print (data)
```

```python
# plotting data
data.plot(x='Hours', y='Scores', style='o')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage') 
plt.title('Graph progression')
plt.show()
```

```python
# Using linear regression algorithm
X = data.iloc[:,0].values.reshape(-1,1)
Y = data.iloc[:,1].values.reshape(-1,1)
```

```python
lr = LinearRegression()
lr.fit(X,Y)
print("Done")
```

```python
# plot the graph
line = lr.coef_*X+lr.intercept_
plt.scatter(X,Y)
plt.show()
```

```python
plt.plot(X, line);
```

```python
plt.scatter(X,Y)
plt.plot(X, line);
```

```python
pred = lr.predict(X) #prediction plot
```

```python
print(pred) #print predicted value
```

```python
plt.scatter(X, pred)
```
